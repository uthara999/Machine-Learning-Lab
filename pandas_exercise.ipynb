```python
null_counts_dict = {}

for column_name in food.columns:
    null_counts_dict[column_name] = food[column_name].isnull().sum()

# Print the dictionary by iterating through its elements using a for loop
for column, null_count in null_counts_dict.items():
    print(f"Column '{column}' has {null_count} null values.")
```
