task5

```python
sns.histplot(x='total_bill', data=df)
plt.show()
```

task6

```python
sns.scatterplot(data=df, x="total_bill", y="tip")

# Set plot title and labels
plt.title("Relationship between Total Bill and Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")

# Display the plot
plt.show()
```

task7

```python

# Create a pairplot using seaborn
sns.pairplot(df)

# Set plot title
plt.suptitle("Relationship between Total Bill, Tip, and Size")

# Display the plot
plt.show()
```

task8

```python
sns.swarmplot(data=df, x="day", y="total_bill")

# Set plot title and labels
plt.title("Relationship between Days and Total Bill")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")

# Display the plot
plt.show()
```

task9

```python
sns.scatterplot(data=df, x="tip", y="day", hue="time")

# Set plot title and labels
plt.title("Relationship between Tip and Day, Differentiated by time")
plt.xlabel("Tip ($)")
plt.ylabel("time")

# Add a legend
plt.legend(title="time")

# Display the plot
plt.show()
```

Task10

```python
sns.boxplot(data=df, x="day", y="total_bill", hue="time")

# Set plot title and labels
plt.title("Total Bill per Day Differentiated by Time")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")

# Add a legend
plt.legend(title="Time")

# Display the plot
plt.show()
```

task11

```python
tips_dinner = df[df["time"] == "Dinner"]
tips_lunch = df[df["time"] == "Lunch"]

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Create histograms using seaborn
sns.histplot(tips_dinner["tip"], ax=axes[0], color="skyblue")
axes[0].set_title("Tip Distribution for Dinner")
axes[0].set_xlabel("Tip ($)")
axes[0].set_ylabel("Frequency")

sns.histplot(tips_lunch["tip"], ax=axes[1], color="salmon")
axes[1].set_title("Tip Distribution for Lunch")
axes[1].set_xlabel("Tip ($)")
axes[1].set_ylabel("Frequency")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

task12

```python
tips_dinner = df[df["time"] == "Dinner"]
tips_Lunch = df[df["time"] == "Lunch"]

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Create scatterplots using seaborn for Male
sns.scatterplot(data=tips_dinner, x="total_bill", y="tip", hue="smoker", ax=axes[0])
axes[0].set_title("Total Bill vs Tip at dinner")
axes[0].set_xlabel("Total Bill ($)")
axes[0].set_ylabel("Tip ($)")
axes[0].legend(title="Smoker")

# Create scatterplots using seaborn for Female
sns.scatterplot(data=tips_Lunch, x="total_bill", y="tip", hue="smoker", ax=axes[1])
axes[1].set_title("Total Bill vs Tip at Lunch")
axes[1].set_xlabel("Total Bill ($)")
axes[1].set_ylabel("Tip ($)")
axes[1].legend(title="Smoker")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
